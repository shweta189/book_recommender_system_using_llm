```python
category_mapping = {'Fiction' : "Fiction",
 'Juvenile Fiction': "Children's Fiction",
 'Biography & Autobiography': "Nonfiction",
 'History': "Nonfiction",
 'Literary Criticism': "Nonfiction",
 'Philosophy': "Nonfiction",
 'Religion': "Nonfiction",
 'Comics & Graphic Novels': "Fiction",
 'Drama': "Fiction",
 'Juvenile Nonfiction': "Children's Nonfiction",
 'Science': "Nonfiction",
 'Poetry': "Fiction"}


books['simple_categories'] = books['categories'].map(category_mapping)
```

```python
books[~books['simple_categories'].isna()]
```

```python
# zero-shot classification

from transformers import pipeline
candidate_labels = ['Fiction','Nonfiction']
classifier = pipeline("zero-shot-classification",model="facebook/bart-large-mnli")
```

```python
sequence_to_classify = books.loc[books['simple_categories']=='Fiction', 'description'].reset_index(drop=True)[0]
```

```python
classifier(sequence_to_classify,candidate_labels)
```

```python
max_index = np.argmax(classifier(sequence_to_classify,candidate_labels)['scores'])
max_label = classifier(sequence_to_classify,candidate_labels)['labels'][max_index]
max_label
```

```python
def generate_prediction(sequence,labels):
  prediction = classifier(sequence,labels)
  max_index = np.argmax(prediction['scores'])
  max_labels = prediction['labels'][max_index]
  return max_labels
```

```python
actual_cls = []
predicted_cls = []
for i in tqdm(range(0,300)):
  sequence = books.loc[books['simple_categories']=='Fiction','description'].reset_index(drop=True)[i]
  predicted_cls += [generate_prediction(sequence,candidate_labels)]
  actual_cls += ['Fiction']

for i in tqdm(range(0,300)):
  sequence = books.loc[books['simple_categories']=='Nonfiction','description'].reset_index(drop=True)[i]
  predicted_cls += [generate_prediction(sequence,candidate_labels)]
  actual_cls += ['Nonfiction']
```

```python
predictions_df = pd.DataFrame({'actual':actual_cls,'predicted':predicted_cls})
```

```python
predictions_df['correct_prediction'] = np.where(predictions_df['actual']==predictions_df['predicted'],1,0)
```

```python
predictions_df['correct_prediction']/len(predictions_df)
```

```python
isbns = []
predicted_cats = []

missing_cats = books.loc[books["simple_categories"].isna(), ["isbn13", "description"]].reset_index(drop=True)
```

```python
for i in tqdm(range(0, len(missing_cats))):
    sequence = missing_cats["description"][i]
    predicted_cats += [generate_prediction(sequence, candidate_labels)]
    isbns += [missing_cats["isbn13"][i]]
```

```python
missing_predicted_df = pd.DataFrame({"isbn13": isbns, "predicted_categories": predicted_cats})
```

```python
books = pd.merge(books, missing_predicted_df, on="isbn13", how="left")
books["simple_categories"] = np.where(books["simple_categories"].isna(), books["predicted_categories"], books["simple_categories"])
books = books.drop(columns = ["predicted_categories"])
```
